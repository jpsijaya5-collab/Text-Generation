```python
import torch
import gradio as gr
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
MODEL_NAME = "gpt2"
```

```python
tokenizer = GPT2Tokenizer.from_pretrained(MODEL_NAME)
model = GPT2LMHeadModel.from_pretrained(MODEL_NAME)
model.to(device)
model.eval()
```

```python
def generate_text(prompt, max_new_tokens, temperature):
   if not prompt.strip():
       return ""

   inputs = tokenizer(prompt, return_tensors="pt").to(device)

   with torch.no_grad():
       outputs = model.generate(**inputs,
                                 max_new_tokens=max_new_tokens,
                                 temperature=temperature,
                                 pad_token_id=tokenizer.eos_token_id)

   generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
   return generated_text
```

```python
with gr.Blocks() as demo:
  gr.Markdown("# GPT-2 Text Generation(Google colab Demo)")
  gr.Markdown("Fast, lightweight, and GPU-aware")

  prompt_input = gr.Textbox(
      label="Prompt",
      placeholder="Once upon a time in a world ruled..."
  )

  with gr.Row():
    max_tokens_slider = gr.Slider(minimum=20, maximum=200, value=80, step=10, label="New tokens")
    temperature_slider = gr.Slider(minimum=0.3, maximum=1.3, value=0.9, step=0.1, label="Temperature")

  generate_button = gr.Button("Generate Text")
  output_text = gr.Textbox(label="Generated Text", interactive=False)

  generate_button.click(
      fn=generate_text,
      inputs=[prompt_input, max_tokens_slider, temperature_slider],
      outputs=output_text
  )

demo.launch(debug=True)
```
